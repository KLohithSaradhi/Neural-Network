```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
```

```python
data
```

```python
data = data.iloc[:,1:]
```

```python
data
```

```python
inputData = data.iloc[:,:-1]
outputData = data.iloc[:,-1:]
```

```python
datain = np.array(inputData)
dataout = np.array(outputData)
```

```python
datain
```

```python
dataout
```

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
```

```python
model = Sequential()
model.add(Dense(2, input_dim = 3, activation = 'linear'))
model.add(Dense(1, activation = 'linear'))

model.compile(loss = "mean_squared_error", optimizer = "adam")
```

```python
model.summary()
```

```python
model.fit(datain, dataout, epochs = 500)

ypred = model.predict(datain)
mse = mean_squared_error(dataout, ypred)
print(mse)
```
