```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data.csv")
```

```python
def normalize(x):
    mean = x.mean()
    std = x.std()
    return (x  - mean)/std
```

```python
x = np.array(data.iloc[:,1])
y = np.array(data.iloc[:,2])
z = np.array(data.iloc[:,3])
out = np.array(data.iloc[:,4])
```

```python
df = {"x" : normalize(x),
     "y" : normalize(y),
     "z" : normalize(z),
     "out" : normalize(out)}
```

```python
df = pd.DataFrame(df)
```

```python
df.to_csv("normalizedData.csv")
```
