```python
import pandas as pd
import numpy as np
```

```python
x = [np.random.randint(100) for i in range(1000)]
y = [np.random.randint(100) for i in range(1000)]
z = [np.random.randint(100) for i in range(1000)]
```

```python
x
```

```python
y
```

```python
z
```

```python
l1_n1_w1 = 0.3
l1_n1_w2 = 0.6
l1_n1_w3 = -0.3

l1_n2_w1 = 0.2
l1_n2_w2 = 0.5
l1_n2_w3 = 0.8

l2_n1_w1 = -0.1
l2_n1_w2 = 0.6

def output(x, y, z):
    return l2_n1_w1 * (l1_n1_w1 * x + l1_n1_w2 * y + l1_n1_w3 * z) + l2_n1_w2 * (l1_n2_w1 * x + l1_n2_w2 * y + l1_n2_w3 * z) 
```

```python
out = [output(x[i], y[i], z[i]) for i in range(1000)]
```

```python
out
```

```python
df = {"x" : x,
     "y" : y,
     "z" : z,
     "out" : out}
```

```python
data = pd.DataFrame(df)
```

```python
data
```

```python
data.to_csv("data.csv")
```
